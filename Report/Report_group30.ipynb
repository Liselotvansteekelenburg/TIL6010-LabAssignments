# 1. Introduction 


The COVID-19 pandemic has significantly affected various sectors, including transportation and mobility. With lockdowns and working remote becomming the norm, travel behaviours have altered, particularly on the dutch highway system. Understanding how these shifts have affected traffic intensities is crucial for future transportation planning and infrastructure development. In the Netherlands, a decline in traffic intensities on highways have become evident. This research examines the extent of the pandemic's impact on traffic intensities, specifically on the A4 highway between Rotterdam and Amsterdam. To address this, one main question and three sub-questions will be explored. 

# 2. Research Objective

The aim of this research is to use traffic data up to 2019 to extrapolate traffic intensities untill 2023. This is then used to show the difference between the extrapolated traffic intensities in 2023 and the measured data. The main objective of this research is to determine the extent to which the COVID-19 pandemic has impacted rush hour traffic intensities on weekdays on the A4 highway between Rotterdam and Amsterdam. The research will look at both directions, right (R) refers to the road from Amsterdam to Rotterdam and left (L) refers to the road of Rotterdam to Amsterdam.

To answer the main objective, three sub questions are introduced.

Main objective: 
#### To what extent did the Covid Pandemic have an impact on rush hour traffic intensities on weekdays on the A4 highway between Rotterdam & Amsterdam?


Sub objectives: 

* What were traffic intensities on the A4 between 2020 and 2023 (after COVID era)?
* What are the projected traffic intensities untill 2023 based on the pre-Covid data? (extrapolated from 2019)
* What is the difference between passenger vehicles and freight traffic?

In doing our research, we have relied on several assumptions to develop a realistic estimate of traffic intensities on the A4. First, we assumed that the traffic data from 2019 represents a normal pre-pandemic condition. This year was chosen as it was the last full calendar year before the disruptions caused by COVID-19.

Additionally, we decided to collect the data from all main trajectories of the A4 to calculate average traffic intensities. This approach was necessary to gain a more comprehensive view of traffic flows, as individual trajectories can exhibit varying intensities due to local factors such as exits, highway intersections and other variables. By utilizing averages, we can better establish the overall trend on the highway and understand the impact of COVID-19 on traffic patterns at a macro level.

We also considered the increasing influence of remote working, which may have resulted in lasting changes to traffic patterns. We assumed that while traffic intensities in 2023 might be lower than the values from 2019, a stabilization could occur depending on how society adapts to new norms. These assumptions form the foundation of our analyses and assist us in interpreting the results within the broader context of the COVID-19 pandemic's impact on traffic intensities.

As lockdowns were implemented and remote work became the norm, the A4 highway experienced drastic changes in traffic intensities. During the early months of the pandemic in 2020, traffic volumes on Dutch highways plummeted. During the weekdays, a minimum was reached in the second week of April, when traffic was only half of what it had been a year earlier (*Centraal Bureau voor de Statistiek,* 2023). The graph shows the traffic intensities during morning (07:00-09:00) and afternoon (16:00-18:00) rush hours on the A4 highway between Rotterdam and Amsterdam for the years 2020 to 2023, showing the effects of the COVID-19 pandemic on traffic patterns.

From 2020 onwards, there is a noticeable trend of gradual recovery in traffic volumes after the initial dip caused by the pandemic in 2020. In 2020, all intensities were significantly lower, which aligns with the global lockdowns and travel restrictions implemented during that period. Morning and afternoon rush hour traffic dropped to roughly 7,000-8,000 vehicles on average per day, indicating a sharp reduction in daily commuting. However, in 2021 and 2022, a clear rebound is observed, with intensities steadily increasing across all categories. This reflects the easing of restrictions and a gradual return to in-person activities, though many workplaces had shifted to hybrid or remote work models, which may have contributed to traffc returning to pre-pandemic levels in 2022. From 2022 onwards the traffic intensities appear to stabilize, with only slight changes compared to previous year. Morning rush hour traffic remains high, especially on the left lane (L), while afternoon rush hour traffic is slightly lower but close to 9,000 vehicles. 


# 4. Extrapolated A4 traffic intensities

The code of this section is in the file 'Ch4.ipynb'.

To accurately extrapolate the pre-COVID traffic data for the A4, key decisions must be made regarding data selection and the type of regression model. In the graph below, traffic intensities from 2012 to 2019 are shown per lane (Left and Right) and for each rush hour (Morning and Afternoon). A significant increase is observed between 2014 and 2015, which can be attributed to the opening of a new section of the A4 between Delft and Schiedam. This expansion increased capacity, leading to a rise in overall traffic intensity. Although the trend continued between 2015 and 2016, the impact of the new road section became less pronounced but still noticeable. 

For extrapolating traffic data to 2023, the period from 2016 to 2019 was selected. A linear regression model was applied to estimate the expected 2023 traffic intensity and compare it with actual 2023 values. 

It is noteworthy that, in the direction towards Amsterdam, the measured traffic intensities for both the morning and afternoon rush hours are lower than the expected values based on the extrapolation. In contrast, for the direction towards Rotterdam, the measured values are higher than the extrapolated intensities. This difference can be explained by several factors:

* Limited dataset for extrapolation: The extrapolation is based on only four years of data (2016-2019). This may result in less reliable predictions. Typically, longer historical periods provide more accurate forecasts.
* Changing travel behavior post-COVID-19: The COVID-19 pandemic has significantly altered travel patterns, particularly for commuting. Many people are now working from home more frequently than before the pandemic, leading to less traffic during peak hours. Additionally, commuters may be choosing to travel outside traditional rush hours, resulting in lower intensities during these times (*Yuan et al.,* 2023).
* ⁠Structural adjustments in infrastructure: Following the opening of the new section of the A4 between Delft and Schiedam, there may have been additional infrastructure changes, such as extra lanes or improved traffic management systems. These modifications could also contribute to the differences between the expected and measured traffic intensities.

The main objective of this study was to assess the accuracy of the extrapolated traffic intensities on the A4, comparing expected and actual values for 2023. The extrapolated data, based on the trend observed from 2016 to 2019, predicted a continued rise in traffic intensity. However, the measured intensities in 2023 showed a deviation from this forecast, with the actual traffic levels returning to values similar to those seen in 2019. This indicates a plateau rather than the expected growth.

Several factors could account for this divergence. First, the post-COVID shift in travel behavior likely played a role, as more individuals work from home and adjust their commuting patterns, choosing to drive outside peak hours. Additionally, the use of a four-year dataset for extrapolation may have introduced limitations in capturing longer-term trends. Consequently, while the forecasted model anticipated a continuation of pre-2019 growth, the actual data reflects a stabilization of traffic intensities, particularly in the direction of Amsterdam, where values fell short of expectations. In contrast, traffic towards Rotterdam exceeded the extrapolated predictions, further underscoring the variability in regional traffic patterns.

Overall, these findings highlight the challenges inherent in relying on short-term trends for long-term forecasting, especially in light of significant external factors like the pandemic.
